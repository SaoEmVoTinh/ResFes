## Bước 1: Mount Google Drive

```
from google.colab import drive
drive.mount('/content/drive')

# Kiểm tra file project có trong Drive chưa
import os
# Đổi đường dẫn nếu bạn để folder tên khác
PROJECT_PATH = '/content/drive/MyDrive/resfes-ar'
print('Files:', os.listdir(PROJECT_PATH))
```

## Bước 2: Copy project vào Colab workspace

```
import shutil
shutil.copytree(PROJECT_PATH, '/content/resfes-ar', dirs_exist_ok=True)
os.chdir('/content/resfes-ar')
print('Working dir:', os.getcwd())
print('Files:', os.listdir('.'))
```

## Bước 3: Cài Buildozer và dependencies

```
%%bash
# Cài system dependencies
apt-get update -qq
apt-get install -qq -y \
    git zip unzip openjdk-17-jdk python3-pip \
    autoconf libtool pkg-config zlib1g-dev \
    libncurses5-dev libncursesw5-dev libtinfo5 \
    cmake libffi-dev libssl-dev

# Cài buildozer
pip install -q buildozer cython==0.29.33
echo '✅ Done'
```

## Bước 4: Build APK (20-30 phút)

```
%%bash
cd /content/resfes-ar
yes | buildozer android debug 2>&1 | tail -100
```

## Bước 5: Kiểm tra và tải APK

```
import glob
apks = glob.glob('/content/resfes-ar/bin/*.apk')
if apks:
    print('✅ APK tạo thành công:')
    for a in apks: print(' ', a)
    
    # Copy về Drive để tải dễ hơn
    import shutil
    for a in apks:
        dest = f'/content/drive/MyDrive/{os.path.basename(a)}'
        shutil.copy(a, dest)
        print(f'Đã copy về Drive: {dest}')
    
    # Hoặc tải thẳng về máy
    from google.colab import files
    files.download(apks[0])
else:
    print('❌ Không tìm thấy APK — xem log ở trên để debug')
```
